# While Loop

while loops continue to perform operations in their block as long as a certain condition is met. For *while* loops to end, the conditional state must be **False** after a while.

                                while (koşul):
                                    İşlem1
                                    İşlem2
                                    İşlem3
                                      //
                                      //


```python
i = 0 

while (i < 10):
    print("i'nin değeri",i)
    i += 1 
```

```python
i = 0 

while (i < 20):
    print("i'nin değeri",i)
    i += 2 
```

```python
i = 0

while (i < 40):
    print(i,"Python Öğreniyorum")
    i +=1
```

```python
liste = [1,2,3,4,5]

a = 0 

while (a < len(liste)):
    print("Indeks:",a,"Eleman:",liste[a])
    a +=1
```
